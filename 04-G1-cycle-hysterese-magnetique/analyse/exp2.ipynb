## Regressions

```python
coefs, cov = np.polyfit(phywe_vide[0], phywe_vide[1], 1, cov=True)
coefserr = unp.uarray(coefs, np.sqrt(cov[(0,1),(0,1)]))
"({:.1uP})x + {:.1uP}".format(*coefserr)
```
