```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from uncertainties import unumpy as unp
from uncertainties import ufloat, umath
from options import *
mpl.rcParams.update(rcParams)

q = -1.602176634e-19  # e-
```

# Ag

I constant = (1.972 $\pm$ 0.001) A

B constant = (343 $\pm$ 6) mT

```python
Ag_B = np.loadtxt("../data/Ag_5branch_B.csv", delimiter=",", skiprows=1)

Bx = Ag_B[1:,0]
Vh = Ag_B[1:,1]

plt.errorbar(Bx, Vh, yerr=0.001 * np.ones(len(Bx)), linestyle='none', capthick=3)
# plt.errorbar(Bx, By, xerr=unp.std_devs(Bx), yerr=unp.std_devs(By))

coefs, cov = np.polyfit(Bx, Vh, 1, cov=True)
fit = np.poly1d(coefs)
coefserr = unp.uarray(coefs, np.sqrt(cov[(0,1),(0,1)]))

plt.plot(Bx, fit(Bx), '--', label=r"$y = ({:.1ueL})x \\+ ({:.1ueL})$".format(*coefserr), color="black")

plt.legend()
plt.xlabel(r"$B$ [\si{\milli \tesla}]")
plt.ylabel(r"$V_H$ [\si{\milli \volt}]")
plt.grid()
plt.tight_layout(pad=0.5)

plt.savefig("../figures/Ag_B.pdf")
```

```python
print(coefserr, "\n")
a = 1.9e-6
Ic = 1.972
deltaIc = 0.001
Rh = -coefs[0] * a / Ic
Rherr = (a/Ic) * 2.0057515567042054e-06 + a*(coefs[0]/(Ic**2))* deltaIc
print(f"{Rh} +- {Rherr}")
```

```python
N = 1/(q * Rh)
deltaN = Rherr/(q * Rh**2)
print(f"{N} +- {deltaN}")
```
