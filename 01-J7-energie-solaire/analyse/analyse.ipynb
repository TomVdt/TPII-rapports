```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
```

## Expérience 2: Puissance max

```python
# Exp 2, d=40cm
amorphe = np.loadtxt("../data/Ir(Ud),amorphe,d=40cm.csv", delimiter=",", skiprows=1, dtype=float)
mono = np.loadtxt("../data/Ir(Ud),monocristallin,d=40cm.csv", delimiter=",", skiprows=1, dtype=float)
poly = np.loadtxt("../data/Ir(Ud),polycristallin,d=40cm.csv", delimiter=",", skiprows=1, dtype=float)
```

```python
plt.scatter(amorphe[:,1], amorphe[:,2])
plt.title("$I_r(U_d), amorphe, d = 40 cm$")
plt.xlabel("$U_d$ [V]")
plt.ylabel("$I_r$ [mA]")
```

```python
plt.scatter(mono[:,1], mono[:,2])
plt.title("$I_r(U_d)$, monocristallin, d = 40 cm")
plt.xlabel("$U_d$ [V]")
plt.ylabel("$I_r$ [mA]")
```

```python
plt.scatter(poly[:,1], poly[:,2])
plt.title("$I_r(U_d)$, polycristallin, d = 40 cm")
plt.xlabel("$U_d$ [V]")
plt.ylabel("$I_r$ [mA]")
```

```python
# Puissances
# [W] = [V] * [A]
puis_amorphe = amorphe[:,1] * amorphe[:,2] * 1e-3

plt.scatter(amorphe[:,1], puis_amorphe)
plt.title("$P(U_d)$, amorphe, d = 40cm")
plt.xlabel("$U_d$ [V]")
plt.ylabel("Puissance [W]")
```

```python
# Puissances
# [W] = [V] * [A]
puis_mono = mono[:,1] * mono[:,2] * 1e-3
plt.scatter(mono[:,1], puis_mono)
plt.title("$P(U_d)$, monocristallin, d = 40cm")
plt.xlabel("$U_d$ [V]")
plt.ylabel("Puissance [W]")
```

```python
# Puissances
# [W] = [V] * [A]
puis_poly = poly[:,1] * poly[:,2] * 1e-3

plt.scatter(poly[:,1], puis_poly)
plt.title("$P(U_d)$, polycristallin, d = 40cm")
plt.xlabel("$U_d$ [V]")
plt.ylabel("Puissance [W]")
```
